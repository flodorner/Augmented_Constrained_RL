```python
import pickle
import numpy as np
from matplotlib import pyplot as plt
```

```python

with open("costs (9).pkl", "rb") as input_file:
    costs = pickle.load(input_file)
with open("rews (9).pkl", "rb") as input_file:
    rewards = pickle.load(input_file)
print(len(costs))

avr_train = []
lowerr_train = []
upperr_train = []
avc_train = []
lowerc_train = []
upperc_train = []
avr_test = []
lowerr_test = []
upperr_test = []
avc_test = []
lowerc_test = []
upperc_test = []

for i in range(30):
    avr_train.append(np.mean(rewards[20*i:20*(i+1)-10]))
    lowerr_train.append(np.mean(rewards[20*i:20*(i+1)-10])-1.96*np.std(rewards[20*i:20*(i+1)-10])/np.sqrt(10))
    upperr_train.append(np.mean(rewards[20*i:20*(i+1)-10])+1.96*np.std(rewards[20*i:20*(i+1)-10])/np.sqrt(10))
    avc_train.append(np.mean(costs[20*i:20*(i+1)-10]))
    lowerc_train.append(np.mean(costs[20*i:20*(i+1)-10])-1.96*np.std(costs[20*i:20*(i+1)-10])/np.sqrt(10))
    upperc_train.append(np.mean(costs[20*i:20*(i+1)-10])+1.96*np.std(costs[20*i:20*(i+1)-10])/np.sqrt(10))
    
    avr_test.append(np.mean(rewards[20*i+10:20*(i+1)]))
    lowerr_test.append(np.mean(rewards[20*i+10:20*(i+1)])-1.96*np.std(rewards[20*i+10:20*(i+1)])/np.sqrt(10))
    upperr_test.append(np.mean(rewards[20*i+10:20*(i+1)])+1.96*np.std(rewards[20*i+10:20*(i+1)])/np.sqrt(10))
    avc_test.append(np.mean(costs[20*i+10:20*(i+1)]))
    lowerc_test.append(np.mean(costs[20*i+10:20*(i+1)])-1.96*np.std(costs[20*i+10:20*(i+1)])/np.sqrt(10))
    upperc_test.append(np.mean(costs[20*i+10:20*(i+1)])+1.96*np.std(costs[20*i+10:20*(i+1)])/np.sqrt(10))
    
x = np.arange(30)*10000

plt.plot(x,avr_train,color="b")
plt.fill_between(x,lowerr_train,upperr_train,alpha=1/2,color="b")
plt.plot(x,avr_test,color="r")
plt.fill_between(x,lowerr_test,upperr_test,alpha=1/2,color="r")
plt.show()

plt.plot(x,avc_train,color="b")
plt.fill_between(x,lowerc_train,upperc_train,alpha=1/2,color="b")
plt.plot(x,avc_test,color="r")
plt.fill_between(x,lowerc_test,upperc_test,alpha=1/2,color="r")
plt.plot(x,np.zeros(30)+25)
plt.show()
```
